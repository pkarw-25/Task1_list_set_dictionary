```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))

# Bar Chart: Total households and owner households over the years
plt.bar(df['year'] - 0.2, df['tot_hhs'], width=0.4, label='Total Households')
plt.bar(df['year'] + 0.2, df['own'], width=0.4, label='Owner Households')
plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Total and Owner Households Over the Years')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['income'], marker='o', label='Average Income')
plt.plot(df['year'], df['expenditure'], marker='o', label='Average Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Average Household Income and Expenditure Over the Years')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))


plt.scatter(df['income'], df['expenditure'], c=df['year'], cmap='viridis', s=100, alpha=0.6, edgecolors='w')
plt.colorbar(label='Year')
plt.xlabel('Average Income ($)')
plt.ylabel('Average Expenditure ($)')
plt.title('Average Household Income vs. Expenditure')

for i in range(len(df)):
    plt.text(df['income'][i], df['expenditure'][i], df['year'][i], fontsize=9, ha='right')

plt.show()
```
