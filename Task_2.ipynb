```python
import pandas as pd
import numpy as np
df=pd.read_csv("01.Data Cleaning and Preprocessing.csv")
```

```python
df.head(10)
```

**Handling Missing Values**

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].apply(lambda col: col.fillna(col.mean()))
```

```python
print(df[numeric_cols].isnull().sum())
```

**Filtering Data Based On Condition**

```python
filtered_df = df[df['Y-Kappa'] > 25]
print(filtered_df)
filtered_df1 = df[df['ChipRate'] > 15]
print(filtered_df1)
```

**Calculating Summary Statistics**

```python
blowflow_mean = df['BlowFlow'].mean()
blowflow_median = df['BlowFlow'].median()
blowflow_std = df['BlowFlow'].std()
```

```python
print("BlowFlow - Mean:", blowflow_mean, "Median:", blowflow_median, "Std:", blowflow_std)
```
